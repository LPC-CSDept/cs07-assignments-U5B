```python
from functools import reduce

# open data from file
caFile = open('ca2021.txt')
flFile = open('fl2021.txt')

# Q1.1
# store input data
caInput = {
  "state": 'CA',
  "male": {
    "gender": 'Male',
    "name": [],
    "number": []
  },
  "female": {
    "gender": 'Female',
    "name": [],
    "number": []
  }
}
flInput = {
  "state": 'FL',
  "male": {
    "gender": 'Male',
    "name": [],
    "number": []
  },
  "female": {
    "gender": 'Female',
    "name": [],
    "number": []
  }
}

# read each line of the file, and split on the spaces
def readLineOfFile(file, state):
  for line in file:
    line = line.strip()
    line = line.split(' ')
    state["male"]["name"].append(line[1])
    state["male"]["number"].append(line[2])
    state["female"]["name"].append(line[3])
    state["female"]["number"].append(line[4])

readLineOfFile(caFile, caInput)
readLineOfFile(flFile, flInput)

# Q1.2
# function to convert gender into dictionary
def genderSort (genderList, outputList, state):
  labels = ["state","gender","number","name"] # this should match the data order
  for num in range(len(genderList["name"])):
    name = genderList["name"][num] # name: 'Neal'
    number = genderList["number"][num] # number of Neal: '1'
    gender = genderList["gender"] # gender: 'Male'
    data = [state]+[gender]+[number]+[name] # this should match label order
    outputList.append(dict(zip(labels,data))) # make a dictionary and zip the data together, then add it to the list
  return outputList

def sortIntoDictionary (file):
  genderList = []
  genderList = genderSort(file["male"], genderList, file["state"]) # fetch males
  genderList = genderSort(file["female"], genderList, file["state"]) # fetch females
  return genderList

caList = sortIntoDictionary(caInput)
flList = sortIntoDictionary(flInput)
statesList = caList + flList # combine the two state lists

# write as JSON data (not in question)
def writeToFile (name, list):
  file = open(name + '.json', 'w+')
  string = str(list)
  string = string.replace('\'', '"')
  file.write(string)
  file.close()
writeToFile('states', statesList)

# Q1.3 & # Q1.4
def findNames (data, state, gender, top, startsWith, accumlative):
  data = list(filter(lambda x: x['state'] == state, data)) # filter by state
  data = list(filter(lambda x: x['gender'] == gender, data)) # filter by gender
  if (startsWith and startsWith != ''): data = list(filter(lambda x: x["name"].startswith(startsWith), data)) # check and see if it starts with input
  data = sorted(data, key=lambda x: x["number"], reverse=True) # sort by number
  if (top and top != 0): data = data[:top] # only include top number of names
  if (accumlative == True): # return accumlative number instead
    data = list(map(lambda x:int(x["number"].replace(",","")), data)) # commas are nasty, get rid of them (also convert it into a integer)
    sumnum = lambda x,y: x+y
    data = reduce(sumnum, data, 0)
  return data

def printAll (list):
  for item in list:
    print(item)

printAll(findNames(statesList, 'CA', 'Male', 10, '', False))
printAll(findNames(statesList, 'FL', 'Female', 10, 'E', False))
print(findNames(statesList, 'FL', 'Female', 10, '', True))
```
