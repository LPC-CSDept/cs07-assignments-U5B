```python

import requests
from bs4 import BeautifulSoup
DRIVER_PATH = '../ignore/chromedriver'
# driver = webdriver.Chrome(executable_path=DRIVER_PATH)
prefixURL = 'https://bw11.clpccd.cc.ca.us/clpccd/2022/01/'
URL = 'https://bw11.clpccd.cc.ca.us/clpccd/2022/02/l/sched_cs.htm'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
# data = requests.get(URL, headers=headers)
# data = data.content
data = open('./sched_cs.htm', 'r', encoding='utf-8').read()

soup = BeautifulSoup(data, 'html.parser')
test = soup.prettify()
table = soup.body.find('table', border="1", cellspacing="0", cellpadding="2", width="100%") # this is the table we want

rows = {}
row_marker = 0
for row in table.find_all('tr'):
  rows[row_marker] = []
  cols = row.find_all('td')
  for col in cols:
    if col.string == None: continue
    if col.string == '\xa0' or col.string == '\xa0 ': 
      if (rows[row_marker] != []): row_marker += 1
      rows[row_marker] = []
      continue
    rows[row_marker].append(col.string)
  if (rows[row_marker] != []): row_marker += 1

for row in rows:
  print(rows[row])
```
